```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import axes3d

font = {'size'   : 16}
plt.rc('font', **font)
```

# Electric field of a uniformly charged rod

The electric field is:
$$\vec{E} = \frac{k_e \lambda}{y} \left[\left(\cos\theta_R - \cos\theta_L\right) \hat{x} + \left(\sin\theta_R - \sin\theta_L\right) \hat{y}\right]$$
where
$$\cos\theta_L = \frac{y}{\sqrt{(-x-d)^2 + y^2}} = \frac{y}{\sqrt{(x+d)^2 + y^2}}$$
$$\cos\theta_R = \frac{y}{\sqrt{(-x+d)^2 + y^2}} = \frac{y}{\sqrt{(x-d)^2 + y^2}}$$
$$\sin\theta_L = \frac{-x-d}{\sqrt{(-x-d)^2 + y^2}} = -\frac{x+d}{\sqrt{(x+d)^2 + y^2}}$$
$$\sin\theta_R = \frac{-x+d}{\sqrt{(-x+d)^2 + y^2}} = -\frac{x-d}{\sqrt{(x-d)^2 + y^2}}$$
note that signs before $x$ and $d$ are there to account for $\theta_L$ or $\theta_R$ being positive or negative.

```python
fig, ax = plt.subplots(figsize=(14, 11));
ax.grid();
ax.set_xlabel("X position", fontsize='large');
ax.set_ylabel("Y position", fontsize='large');

size = 4;
nrvec = [20,20];

l = 5.0;
d = l/2;
lam_ke = 1.0;
x,y = np.meshgrid(np.linspace(-size*d, size*d, nrvec[0]),np.linspace(-size*d, size*d, nrvec[1]))

Ex = (lam_ke/y)*(y/np.sqrt(np.power(x-d, 2) + np.power(y, 2)) - y/np.sqrt(np.power(x+d, 2) + np.power(y, 2)));
Ey = -(lam_ke/y)*((x-d)/np.sqrt(np.power(x-d, 2) + np.power(y, 2)) - (x+d)/np.sqrt(np.power(x+d, 2) + np.power(y, 2)));
E_tot = np.sqrt(np.power(Ex, 2) + np.power(Ey, 2));

ax.plot([-d, d], [0., 0.], '-', c='black', lw=5);
emfield_rod = ax.streamplot(x, y, Ex, Ey, color=E_tot, linewidth=2, cmap="rainbow",
                            density=1.6, arrowstyle='->', arrowsize=1.5);
fig.colorbar(emfield_rod.lines);
plt.savefig("Efield_uniform_rod.png", dpi=200, facecolor='w', bbox_inches="tight");
```

# Electric field of a uniformly charged ring, side view

In a side view, this is the same as the electric field of two equal point charges:

The electric field is:
$$\vec{E} = k_e Q \left[\left(\sin\theta_1 + \sin\theta_2\right) \hat{x} + \left(\cos\theta_1 + \cos\theta_2\right) \hat{y}\right]$$
where
$$\sin\theta_1 = \frac{x+R}{\left((x+R)^2 + y^2\right)^{3/2}}$$
$$\sin\theta_2 = \frac{x-R}{\left((x-R)^2 + y^2\right)^{3/2}}$$
$$\cos\theta_1 = \frac{y}{\left((x+R)^2 + y^2\right)^{3/2}}$$
$$\cos\theta_2 = \frac{y}{\left((x-R)^2 + y^2\right)^{3/2}}$$

```python
fig, ax = plt.subplots(figsize=(14, 11));
ax.grid();
ax.set_xlabel("X position", fontsize='large');
ax.set_ylabel("Y position", fontsize='large');

size = 4;
nrvec = [20,20];

R = 2.5;
Q_ke = 1.0;
x,y = np.meshgrid(np.linspace(-size*R, size*R, nrvec[0]),np.linspace(-size*R, size*R, nrvec[1]));

Ex = Q_ke*((x+R)/np.power(np.power(x+R, 2) + np.power(y, 2), 3/2) + (x-R)/np.power(np.power(x-R, 2) + np.power(y, 2), 3/2));
Ey = Q_ke*(y/np.power(np.power(x+R, 2) + np.power(y, 2), 3/2) + y/np.power(np.power(x-R, 2) + np.power(y, 2), 3/2));
E_tot = np.sqrt(np.power(Ex, 2) + np.power(Ey, 2));

ax.plot([-R, R], [0., 0.], '-', c='black', lw=5);
emfield_ring = ax.streamplot(x, y, Ex, Ey, color=E_tot, linewidth=2, cmap="rainbow",
                             density=1.6, arrowstyle='->', arrowsize=1.5);
fig.colorbar(emfield_ring.lines);
plt.savefig("Efield_uniform_ring_side.png", dpi=200, facecolor='w', bbox_inches="tight");
```
